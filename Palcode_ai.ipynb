```python
yaml_text = """
train: /content/drive/MyDrive/palcode.ai/images/train
val: /content/drive/MyDrive/palcode.ai/images/val

names:
  0: window
  1: door
"""
with open("/content/drive/MyDrive/palcode.ai/data.yaml", "w") as f:
    f.write(yaml_text)
```

```python
from ultralytics import YOLO

model = YOLO('yolov8n.pt')  # YOLOv8 Large

model.train(data='/content/drive/MyDrive/palcode.ai/data.yaml', epochs=50, imgsz=640)
```

```python
model = YOLO('/content/runs/detect/train/weights/best.pt')

results = model.predict(source='/content/drive/MyDrive/palcode.ai/images/k.webp', save=True, conf=0.3)
```
